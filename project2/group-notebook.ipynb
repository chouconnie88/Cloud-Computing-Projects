```
# EXAMPLE

# Analyzing the dataset for potential signs of bot activity

# Checking the number of changes made by each user
user_activity = maddie_sample['user'].value_counts()

# Checking for rapid succession of actions by looking at timestamp differences
maddie_sample['timestamp'] = pd.to_datetime(maddie_sample['timestamp'])
maddie_sample = maddie_sample.sort_values(by='timestamp')
maddie_sample['time_diff'] = maddie_sample['timestamp'].diff().dt.total_seconds().fillna(0)

# Top 10 users with the most activity
top_users = user_activity.head(10)

# Average time difference for changes made by the top 10 most active users
avg_time_diff_top_users = maddie_sample[maddie_sample['user'].isin(top_users.index)].groupby('user')['time_diff'].mean()

# Display the results
top_users, avg_time_diff_top_users
```

1. User Activity Volume: The top 10 most active users have a significant number of actions, with the highest being 55 actions. This high level of activity might suggest the use of bots, especially if these actions were sustained over a short period.

2. Average Time Difference for Top Users: The average time difference between actions for these top users is very low, ranging from about 0.15 to 0.22 seconds. This rapid succession of actions is faster than typical human behavior and strongly suggests bot activity.

These findings indicate that certain users might have used bots to automate their actions in the r/place dataset. The extremely short intervals between actions and the high volume of activity are key indicators of such automated behavior. ​​

Methodology for how to determine what artwork was most affected by bots:

1. Identify High Activity Areas: First, we need to find which areas of the canvas had the most activity. This can be done by analyzing the frequency of pixel changes in different coordinates.

2. Analyze Specific Users: Focus on the users who were identified as potentially being bots based on their high activity and rapid succession of actions. Examine the coordinates and colors they were changing.

3. Pattern Recognition: Look for patterns in the pixel changes. Bots are likely to create repetitive patterns or maintain certain images or colors consistently.

4. Temporal Analysis: Analyze the timestamps to see if there are specific time periods where bot activity peaked, and correlate this with changes in certain areas of the canvas.

5. Visualize the Data: Create visualizations to map out where the most changes occurred and which users were involved in these changes. This can help identify which artworks were most affected by bots.

```
# Step 1: Identify High Activity Areas

# Filtering out rows with non-standard coordinate formats
standard_format_rows = combined_df['coordinate'].str.contains(r'^-?\d+,-?\d+$')
filtered_data = combined_df[standard_format_rows]

# Repeating the coordinate splitting and conversion to numeric values
coordinates_filtered = filtered_data['coordinate'].str.split(',', expand=True)
filtered_data['x'] = pd.to_numeric(coordinates_filtered[0])
filtered_data['y'] = pd.to_numeric(coordinates_filtered[1])

# Calculating the frequency of pixel changes for each coordinate again
pixel_change_frequency_filtered = filtered_data.groupby(['x', 'y']).size().reset_index(name='change_count')

# Sorting by change_count to find the most frequently changed pixels
high_activity_areas_filtered = pixel_change_frequency_filtered.sort_values(by='change_count', ascending=False).head()

high_activity_areas_filtered
```

```
# Step 2: Analyze Specific Users (Potential Bots)

# Identifying users with high activity
# We consider a user as highly active if they have made changes more than a certain threshold. 
# Here, we choose a threshold (e.g., top 5% active users), but this can be adjusted based on the dataset characteristics.
activity_threshold = filtered_data['user'].value_counts().quantile(0.95)
high_activity_users = filtered_data['user'].value_counts()[filtered_data['user'].value_counts() > activity_threshold].index

# Filtering data for high activity users
bot_like_activity = filtered_data[filtered_data['user'].isin(high_activity_users)]

# Analyzing the coordinates and colors changed by these users
bot_activity_analysis = bot_like_activity.groupby(['user', 'x', 'y', 'pixel_color']).size().reset_index(name='change_count')

# Sorting to see the most frequent changes by these users
bot_activity_analysis_sorted = bot_activity_analysis.sort_values(by='change_count', ascending=False).head()

bot_activity_analysis_sorted
```

These users exhibit a high frequency of changes at specific coordinates, suggesting focused activity on certain parts of the canvas. The repetition in changing pixels to specific colors indicates a pattern, possibly for creating or altering specific imagery. Some users appear to be collaborating or competing, as indicated by changes to similar coordinates (e.g., (136, 216)) by different users. The high number of changes (8-10 times) at the same coordinates could imply automated or bot-driven behavior, especially if these changes occurred within a short time frame.

```
import matplotlib.pyplot as plt

# Identifying repeated colors (assuming 'pixel_color' is a column in the dataset)
repeated_colors = bot_like_activity['pixel_color'].value_counts().head()

# Identifying repeated coordinates (assuming 'x' and 'y' are columns in the dataset)
repeated_coordinates = bot_like_activity.groupby(['x', 'y']).size().reset_index(name='change_count')
repeated_coordinates_sorted = repeated_coordinates.sort_values(by='change_count', ascending=False).head()

# Plotting the bar graph for repeated colors
plt.figure(figsize=(10, 6))
repeated_colors.plot(kind='bar')
plt.title('Top Repeated Colors')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plotting the bar graph for repeated coordinates
plt.figure(figsize=(10, 6))
repeated_coordinates_sorted.plot(kind='bar', x='x', y='change_count')
plt.title('Top 5 Repeated Coordinates from Top 5% Active Users')
plt.xlabel('(X, Y) Coordinates')
plt.ylabel('Change Count')
plt.xticks(rotation=45)
plt.show()

repeated_colors, repeated_coordinates_sorted
```

```
# Step 3: Pattern Recognition

# Analyzing for repetitive patterns in the changes made by high activity users
# We look for patterns such as repeated colors, coordinates, or specific sequences of changes

# Identifying repeated colors
repeated_colors = bot_like_activity['pixel_color'].value_counts().head()

# Identifying repeated coordinates
repeated_coordinates = bot_like_activity.groupby(['x', 'y']).size().reset_index(name='change_count')
repeated_coordinates_sorted = repeated_coordinates.sort_values(by='change_count', ascending=False).head()

# Displaying the top repeated colors and coordinates
repeated_colors, repeated_coordinates_sorted
```

The dominance of certain colors like white, black, and orange suggests that these colors are preferred for certain patterns or images, possibly due to their visibility or symbolic meaning. The high frequency of changes at specific coordinates indicates concentrated activity in these areas, which could be areas of artistic focus, contention, or collaboration. The extreme repetition at certain coordinates (over 100 times) strongly implies automated processes or a concerted effort by multiple users to maintain or alter the appearance at these spots.

```
import pandas as pd

# Step 4: Temporal Analysis and Step 5: Visualization

# Converting 'timestamp' to datetime format for analysis
filtered_data['timestamp'] = pd.to_datetime(filtered_data['timestamp'])

# Analyzing the distribution of changes over time
# Grouping data by time intervals (e.g., hourly) and counting changes
time_grouped_activity = filtered_data.set_index('timestamp').groupby(pd.Grouper(freq='H')).size()

# Plotting the distribution of changes over time
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
time_grouped_activity.plot(kind='bar')
plt.title('Distribution of Pixel Changes Over Time')
plt.xlabel('Time')
plt.ylabel('Number of Changes')
plt.xticks(rotation=45)
plt.show()
```
