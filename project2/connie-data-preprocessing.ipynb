```python
#create a function to convert timestamp column to datetime for each file put in
def preprocessing(url):
    
    #upload csv file
    df = pd.read_csv(url)
    
    #change timestamp to datetime
    df["timestamp"] = pd.to_datetime(df["timestamp"], errors='coerce')
    
    #systematic sampling to get every 50 rows, including the first row
    sampling_rate = 50

    sampled_df_name = df.iloc[::sampling_rate]

    # Print the sampled DataFrame
    return sampled_df_name

```

```python
#concat all of the samples together
cc_sample = pd.concat([zero_sample, one_sample, two_sample, three_sample, four_sample, five_sample,
                      six_sample, seven_sample, eight_sample, nine_sample, ten_sample, eleven_sample,
                      twelve_sample, thirteen_sample, fourteen_sample, fifteen_sample, sixteen_sample,
                      seventeen_sample])
cc_sample
```

```python
#turn concatenated stuff into csv
# cc_sample.to_csv(header=True, compression='gzip')

# Define the chunk size (e.g., 10,000 rows)
chunk_size = 10000

# Create an empty list to store the chunks
chunks = []

# Load your large DataFrame into memory or create it if it's not already loaded

# Calculate the number of chunks needed to process your entire DataFrame
num_chunks = len(cc_sample) // chunk_size + 1

# Create a loop to iterate through the chunks and append them to the list
for i, chunk in enumerate(np.array_split(cc_sample, num_chunks)):
    chunks.append(chunk)

# Concatenate all the chunks into a single DataFrame
result_df = pd.concat(chunks, ignore_index=True)

# Save the result DataFrame as a single CSV file
result_df.to_csv('combined_data.csv', index=False)
```
